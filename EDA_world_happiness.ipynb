## Проанализируем Доклад о мировом счастье за 2018-2019 года.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import legend
```

```python
# увеличим размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 11,7
```

```python
df_2019 = pd.read_csv('2019.csv')
df_2019
```

Датасет состоит из 9 колонок - Общий рейтинг, Страна или регион, ВВП на душу населения, Социальная поддержка, Ожидаемая продолжительность жизни, Свобода делать жизненный выбор, Щедрость, Восприятие коррупции.

```python
# Удалим столбец Overall rank, так как полезность его 0

df_2019 = df_2019.drop('Overall rank', axis = 1)
```

```python
# Изменим название столбца 'Country or region'на 'Country'

df_2019.rename(columns={'Country or region':'Country'}, inplace=True)
```

```python
df_2019.info()
print('-----------------------------------------------------------------')
df_2019.describe()
```

```python
df_2018 = pd.read_csv('2018.csv')
df_2018.head()
```

```python
# Удалим столбец Overall rank, так как полезность его 0

df_2018 = df_2018.drop('Overall rank', axis = 1)
```

```python
# Изменим название столбца 'Country or region'на 'Country'

df_2018.rename(columns={'Country or region':'Country'}, inplace=True)
```

```python
df_2018.info()
print('-----------------------------------------------------------------')
df_2018.describe()
```



Данные не содержат пропусков, типы данных корректны.

```python
# Построим гистограмму для визуализации распределения значений по каждому столбцу за 2018 год
df_2018.hist(bins=10)
plt.tight_layout()
```

```python
# Построим гистограмму для визуализации распределения значений по каждому столбцу за 2019 год
df_2019.hist(bins=10)
plt.tight_layout()
```

При сравнении 2018 и 2019 видно, как изменилось распределение по столбцам: Score, GDP per capita, Healthy life expectancy, Generosity.

Далее подробнее изучим некоторые из них.

```python
# Сравним распределение уровня счастья за 2018 и 2019г

df_2018['Score'].plot(kind = 'hist', color= '#FF1800',
                     alpha = 0.5,
                    label = '2018',
                    density = True)
df_2019['Score'].plot(kind = 'hist', color= '#00C12B',
                     alpha = 0.5,
                    label = '2019',
                    density = True)
plt.legend(loc='best')
plt.title('Сравнение распределений рейтинга уровня счастья')
```

Наибольший количество стран с уровнем счастья в диапазоне 4.5 - 6. В 2018 наблюдаеются две моды, которые в 2019 еще больше выросли. 30% стран с уровнем счастья ниже 4.5 и 30% - выше 6.

```python
df_2018['GDP per capita'].plot(kind = 'hist', color = '#1F1AB2',
                     alpha = 0.5,
                    label = '2018',
                    density = True)
df_2019['GDP per capita'].plot(kind = 'hist',color = '#FFC200',
                     alpha = 0.5,
                    label = '2019',
                    density = True)
plt.legend(loc='best')
plt.title('Сравнение распределений ВВП на 1 человека')
```

В 2018 году показатель постепенно увеличивается, на пике около 1.0 начинает падать и на показателе 1.5 резкий спад. В 2019 вершина сдвигается вправо, то есть доля стран с высоким значением ВВП увеличивается, но максимальное значение ВВП на 1 человека изменилось с 2.0 на 1.7.

```python
df_2018['Healthy life expectancy'].plot(kind = 'hist', color = '#0772A1',
                     alpha = 0.5,
                    label = '2018',
                    density = True)
df_2019['Healthy life expectancy'].plot(kind = 'hist', color = '#FF8700',
                     alpha = 0.5,
                    label = '2019',
                    density = True)
plt.legend(loc='best')
plt.title('Сравнение распределений продолжительности здоровой жизни')
```

В 2018 году постепенное увеличение показателя, на пике 0.6 уменьшается. В 2019 году гистограмма сдвигается вправо, и мода теперь равна 0.9, то есть количество стран с наибольшим значение увеличивается - увеличивается продолжительность жизни, а с наименьшим значением - количество уменьшается.

```python
# Топ 5 стран с наибольшим ВВП на 1 человека в 2018 и 2019 годах

df_2018_top5_GDP = df_2018[['Country', 'GDP per capita']].sort_values('GDP per capita', ascending = False).head(5)
df_2018_top5_GDP.plot(x ='Country',
                    kind='barh',
                    title='Топ 5 стран с высоким ВВП на 1 человека в 2018 году',
                    xlabel='Страны',
                    ylabel='ВВП на 1 человека',
                    legend = False)
plt.xlim(0,2.2)
df_2019_top5_GDP = df_2019[['Country', 'GDP per capita']].sort_values('GDP per capita', ascending = False).head(5)
df_2019_top5_GDP.plot(x ='Country',
                    kind='barh',
                    title='Топ 5 стран с высоким ВВП на 1 человека в 2019 году',
                    xlabel='Страны',
                    ylabel='ВВП на 1 человека',
                    legend = False)
plt.xlim(0,2.2)
```

В 2019 году в пятерку вошли те же страны, что и в 2018 году, только изменился рейтинг - Арабские Эмираты с 1-го места сместились на 4-е, их позицию заняла арабская страна Катар. Сингапур с 4 места поднялся на 3. Люксембург с 3 - на 2. Кувейт остался на 5 месте. Изменилось значение ВВП только у  АЭ -  с 2.1 упало до 1.6, у остальных сильных измений нет.

Самое интересное, что эти страны не попали в топ-5 по уровню счастья.

Проверим шесть показателей на взаимосвязь с уровнем счастья. И попробуем определить, что влияет на счастье.

```python
# Есть ли взаимосвязь между уровнем счастья и значением ВВП на 1 человека?

sns.regplot(data=df_2018, x='GDP per capita', y='Score', color='green', marker='*')
plt.ylim(0, 9)

sns.regplot(data=df_2019, x='GDP per capita', y='Score', color='red', marker='+')
plt.title('Зависимость уровня счастья и ВВП на 1 человека')
plt.ylim(0, 9)
plt.text(0.00, 8, '* 2018 год', color='green')
plt.text(0.00, 7.5, '+ 2019 год',color='red' )
plt.text(0.5,2, 'Прослеживается взаимосвязь - с ростом ВВП уровень счастья выше.')
plt.text(0.5,1.6, 'Одинаково в 2018 и 2019 годах')
```

```python
# Есть ли взаимосвязь между уровнем счастья и социальной поддержкой?

sns.regplot(data=df_2018, x='Social support', y='Score', color='green', marker='*')
plt.ylim(0, 9)

sns.regplot(data=df_2019, x='Social support', y='Score', color='red', marker='+')
plt.title('Зависимость уровня счастья и социальной поддержкой')
plt.ylim(0, 9)
plt.text(0.00, 8, '* 2018 год', color='green')
plt.text(0.00, 7.5, '+ 2019 год',color='red' )
plt.text(0.5,2, 'Прослеживается взаимосвязь - с ростом значения - уровень счастья выше.')
plt.text(0.5,1.6, 'Одинаково в 2018 и 2019 годах')
```

```python
# Есть ли взаимосвязь между уровнем счастья и продолжительностью жизни?

sns.regplot(data=df_2018, x='Healthy life expectancy', y='Score', color='green', marker='*')
plt.ylim(0, 9)

sns.regplot(data=df_2019, x='Healthy life expectancy', y='Score', color='red', marker='+')
plt.title('Зависимость уровня счастья и продолжительностью жизни')
plt.ylim(0, 9)
plt.text(0.00, 8, '* 2018 год', color='green')
plt.text(0.00, 7.5, '+ 2019 год',color='red' )
plt.text(0.4,2, 'Взаимосвязь проявляется. Линия регрессии и значения в 2019 году')
plt.text(0.4,1.6, 'опустились по оси у. Это значит, чем выше уровень счастья, тем выше')
plt.text(0.4,1.2, 'продолжительность жизни в сравнении с 2018 годом.')
```

```python
# Есть ли взаимосвязь между уровнем счастья и свободой делать жизненный выбор?

sns.regplot(data=df_2018, x='Freedom to make life choices', y='Score', color='green', marker='*')
plt.ylim(0, 9)

sns.regplot(data=df_2019, x='Freedom to make life choices', y='Score', color='red', marker='+')
plt.title('Зависимость уровня счастья и свободой')
plt.ylim(0, 9)
plt.text(0.00, 8, '* 2018 год', color='green')
plt.text(0.00, 7.5, '+ 2019 год',color='red' )
plt.text(0.35,2, 'Взаимосвязь есть, но слабая.')
plt.text(0.35,1.6, 'Значения 2018 и 2019 года более хаотичны.')
```

```python
# Есть ли взаимосвязь между уровнем счастья и щедростью, великодушием?

sns.regplot(data=df_2018, x='Generosity', y='Score', color='green', marker='*')
plt.ylim(0, 9)

sns.regplot(data=df_2019, x='Generosity', y='Score', color='red', marker='+')
plt.title('Зависимость уровня счастья и щедростью')
plt.ylim(0, 9)
plt.text(0.00, 8.5, '* 2018 год', color='green')
plt.text(0.00, 8, '+ 2019 год',color='red' )
plt.text(0.26,2, 'Взаимосвязи нет. Значения равномерно распределены')
plt.text(0.26,1.6, 'между интервалом от 0.0 до 0.4 как в 2018 году, так и в 2019.')
```

```python
# Есть ли взаимосвязь между уровнем счастья и восприятием коррупции?

sns.regplot(data=df_2018, x='Perceptions of corruption', y='Score', color='green', marker='*')
plt.ylim(0, 9)

sns.regplot(data=df_2019, x='Perceptions of corruption', y='Score', color='red', marker='+')
plt.title('Зависимость уровня счастья и восприятием коррупции')
plt.ylim(0, 9)
plt.text(0.00, 8.5, '* 2018 год', color='green')
plt.text(0.00, 8, '+ 2019 год',color='red' )
plt.text(0.1,2, 'Большая часть значений (до 0.2 по оси х) распределены равномерно,')
plt.text(0.1,1.6, 'а остальная - с увеличением показателя также растет уровень счастья.')
```

На графиках видно, что на уровень счастья влияют ВВП на 1 человека, социальная поддержка и продолжительность жизнь, небольшая зависимость есть с показателем - свобода делать жизненный выбор. А щедрость и восприятие коррупции на уровень счастья влияют меньше всего, можно сказать, не влияют.

```python
# Проверим как связаны между собой значения, у которых нашли взаимосвязь с уровнем счастья
df_18 = df_2018[['GDP per capita', 'Social support','Healthy life expectancy', 'Freedom to make life choices']]

sns.pairplot(df_18)
```

```python
df_19 = df_2019[['GDP per capita', 'Social support','Healthy life expectancy', 'Freedom to make life choices']]

sns.pairplot(df_19)
```

Видна взаимосвязь между собой таких показателей, как ВВП, социальная поддержка и продолжительность жизни в 2018 и 2019 годах. Свобода больше взаимосвязана с уровнем счастья, чем с экономическими показателями.

```python

plt.subplot(1,2,1)
sns.boxplot(data = df_2018['Score'])
plt.title('Уровень счастья 2018 год')

plt.subplot(1,2,2)
sns.boxplot(data = df_2019['Score'])
plt.title('Уровень счастья 2019 год')
```

Как видно на визуализации, значения 2018 и 2019 года не изменились. Совсем незначительно уменьшился межквартильный размах в 2019 году, первый квантиль стал немного больше, а третий квантиль - немного меньше.

```python
plt.subplot(1,2,1)
sns.boxplot(data = df_2018[['GDP per capita', 'Social support','Healthy life expectancy', 'Freedom to make life choices',
                            'Generosity', 'Perceptions of corruption']])
plt.title('2018 год')
plt.xticks(rotation=90)
plt.ylim(0, 2.2)

plt.subplot(1,2,2)
sns.boxplot(data = df_2019[['GDP per capita', 'Social support','Healthy life expectancy', 'Freedom to make life choices',
                            'Generosity', 'Perceptions of corruption']])
plt.title('2019 год')
plt.xticks(rotation=90)
plt.ylim(0, 2.2)
```

Показатель "Восприятие коррупции" имеет наибольшее количесво выбросов - это страны с высоким уровнем доверия к правительству, то есть с низкой коррупцией.
Продолжительность жизни - значения в 2019 году увеличились (минимум, медиана, максимум). А значения Свободы, наоборот, незначительно снизились.

Вывод.

На уровень счастья влияют  такие показатели, как ВВП на 1 человека, социальная поддержка и продолжительность жизнь, небольшая взаимосвязь есть с показателем - свобода делать жизненный выбор. Но это не означает 100% влияние. Так как, например, топ-5 стран с высоким ВВП на 1 человека, не попали в пятерку счастливых стран. Поэтому на счастье влияют еще много факторов, которых нет в исследуемых данных.
